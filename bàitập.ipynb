```python
import pandas as pd
import numpy as np

# 3 file CSV nằm cùng thư mục với notebook
tep_info = 'sensor_info.csv'
tep_readings = 'sensor_readings.csv'
tep_maint = 'maintenance_log.csv'
```

```python
df_thong_tin = pd.read_csv(tep_info)
df_doc_cam_bien = pd.read_csv(tep_readings)
df_bao_tri = pd.read_csv(tep_maint)

print('=== sensor_info ===')
print(df_thong_tin.head())
print('\n=== sensor_readings ===')
print(df_doc_cam_bien.head())
print('\n=== maintenance_log ===')
print(df_bao_tri.head())
```

```python
# Chuẩn hóa sensor_id: bỏ khoảng trắng, viết hoa
df_thong_tin['sensor_id'] = (
    df_thong_tin['sensor_id']
    .astype(str)
    .str.strip()
    .str.upper()
)

# Chuẩn hóa machine_id: bỏ khoảng trắng, viết hoa
df_thong_tin['machine_id'] = (
    df_thong_tin['machine_id']
    .astype(str)
    .str.strip()
    .str.upper()
)

# Chuẩn hóa location
loc_raw = df_thong_tin['location'].astype(str).str.strip()
loc_key = loc_raw.str.lower().str.replace(' ', '')

# Line 1 / Line1 / line 1 -> 'Line 1'
df_thong_tin.loc[loc_key.str.contains('line1'), 'location'] = 'Line 1'

# khu a / Khu  A / Khu A -> 'Khu A'
df_thong_tin.loc[loc_key.str.contains('khua'), 'location'] = 'Khu A'

# Các trường hợp khác: viết hoa chữ cái đầu
mask_other = ~(loc_key.str.contains('line1') | loc_key.str.contains('khua'))
df_thong_tin.loc[mask_other, 'location'] = loc_raw[mask_other].str.title()

# Chuẩn hóa sensor_type
stype_raw = df_thong_tin['sensor_type'].astype(str).str.strip()
stype_low = stype_raw.str.lower()

df_thong_tin.loc[stype_low.str.contains('temp'), 'sensor_type'] = 'Temperature'
df_thong_tin.loc[stype_low.str.contains('vib'), 'sensor_type'] = 'Vibration'
df_thong_tin.loc[stype_low.str.contains('curr'), 'sensor_type'] = 'Current'

df_thong_tin.head()
```
